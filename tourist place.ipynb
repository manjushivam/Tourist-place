```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('Top Indian Places to Visit.csv')
col=['Type','Google review rating','Number of google review in lakhs','Entrance Fee in INR']
df=df[col]
```

```python
df = df[df['Entrance Fee in INR'] != 0]
print(df)
df.dropna(inplace=True)
```

```python
df=pd.get_dummies(df,dtype=int)
df
```

```python
X=df.drop(columns=['Entrance Fee in INR'])
Y=df['Entrance Fee in INR']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.40)
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
print(model.score(X_train,Y_train))
print(model.score(X_test,Y_test))
```

```python
from sklearn.linear_model import Lasso
model2=Lasso(alpha=20,max_iter=100,tol=0.3)
model2.fit(X_train,Y_train)
```

```python
print(model2.score(X_test,Y_test))
```

```python
from sklearn.linear_model import Ridge
model3=Ridge(alpha=0.1,max_iter=100,tol=0.1) #alpha &max increased -> accuracy decreased
model3.fit(X_train,Y_train)
```

```python
print(model3.score(X_test,Y_test))
```
